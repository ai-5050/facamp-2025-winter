Reference: https://python.langchain.com/api_reference/community/document_loaders/langchain_community.document_loaders.web_base.WebBaseLoader.html#webbaseloader

```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
# 바르셀로나, 벤피카에 5-4 대역전승…UCL 16강 진출
url = "https://www.newsis.com/view/NISX20250122_0003041476#"
```

## Crawling

### Entire Page

```python
loader = WebBaseLoader(web_path=url, verify_ssl=False)
```

```python
page = loader.load()
```

```python
page
```

```python
len(page)
```

```python
page[0].metadata
```

```python
print(page[0].page_content)
```

### Only Content

```python
# pip install beautifulsoup4==4.13.1
import bs4
```

```python
# using BeautifulSoup
loader = WebBaseLoader(
    web_path=url,
    verify_ssl=False,
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("view")
        )
    ),
)
```

```python
page = loader.load()
```

```python
page
```

```python
len(page)
```

```python
print(page[0].page_content)
```

## Splitting

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,
    chunk_overlap=0
)
```

```python
texts = text_splitter.split_documents(page)
```

```python
texts
```

```python
len(texts)
```

## Embedding

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embedding_model = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
text_content = [text.page_content for text in texts]
```

```python
text_content
```

```python
embeddings = embedding_model.embed_documents(text_content)
```

```python
len(embeddings)
```
