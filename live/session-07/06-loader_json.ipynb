````python
# define the metadata extraction function.
def metadata_func(record: dict, metadata: dict) -> dict:
    metadata["source"] = record["metadata"]["source"]
    metadata["title"] = record["metadata"]["title"]
    return metadata

loader = JSONLoader(
    file_path="./data/news-sample.json",
    jq_schema=".[]",
    content_key="page_content",
    metadata_func=metadata_func
)
````

````python
docs = loader.load()
````

## Tokens

````python
import tiktoken
````

````python
# get the tokeniser corresponding to a specific model in the OpenAI API
encoder = tiktoken.encoding_for_model("gpt-4o-mini")
````

````python
for doc in docs:
    tokens = encoder.encode(doc.page_content)
    print("Tokens:", len(tokens), "/ Title:", doc.metadata["title"])
````

## Splitting

Not really necessary because the number of tokens for each document is within the context window.

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter
````

````python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=3500,
    chunk_overlap=0
)
````

````python
docs_splitted = text_splitter.split_documents(docs)
````

````python
docs_splitted
````

````python
len(docs_splitted)
````

## Embedding

````python
from langchain_openai import OpenAIEmbeddings
````

````python
embedding_model = OpenAIEmbeddings(model="text-embedding-3-small")
````

````python
doc_content = [doc.page_content for doc in docs_splitted]
````

````python
doc_content
````

````python
embeddings = embedding_model.embed_documents(doc_content)
````
